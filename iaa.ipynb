```python
from google.colab import drive
import os


drive.mount('/content/drive', force_remount=True)
folder_path = '/content/drive/My Drive/MSc Computer Science/Master Thesis/Annotations/'

for f in os.listdir(folder_path):
    print(f)
```

```python
import pandas as pd

author_anno_fn = 'agreements_author.xlsx'
second_anno_fn = 'second_agreements.xlsx'

df_author = pd.read_excel(folder_path + author_anno_fn)
df_second = pd.read_excel(folder_path + second_anno_fn)
```

```python
df_author
```

```python
df_second
```

```python
aut_anno = df_author['anno'].tolist()
sec_anno = df_second['anno'].tolist()
```

```python
from sklearn.metrics import cohen_kappa_score

kappa_quadratic = cohen_kappa_score(aut_anno, sec_anno, weights='quadratic')
print(f"Quadratic weighted kappa: {kappa_quadratic:.3f}")
```

# Diachronic

```python
from sklearn.metrics import cohen_kappa_score

d_aut_r1 = df_author[(df_author['corpora'] == 'D')]['anno'][:5].tolist()
d_sec_r1 = df_second[(df_second['corpora'] == 'D')]['anno'][:5].tolist()
print('kappa dia r1', cohen_kappa_score(d_aut_r1, d_sec_r1, weights='quadratic'))

d_aut_tot = df_author[df_author['corpora'] == 'D']['anno'].tolist()
d_sec_tot = df_second[df_second['corpora'] == 'D']['anno'].tolist()
print('kappa dia total', cohen_kappa_score(d_aut_tot, d_sec_tot, weights='quadratic'))
```
